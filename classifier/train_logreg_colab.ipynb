```python
import json
import joblib
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt

okt = Okt()

DATA_PATH = "data/train_data.json"
MODEL_DIR = "models"
os.makedirs(MODEL_DIR, exist_ok=True)


def extract_nouns(text):
    return " ".join(okt.nouns(text))


def load_data():
    with open(DATA_PATH, "r", encoding="utf-8") as f:
        data = json.load(f)
    texts = [d["text"] for d in data]
    return texts


print("데이터 로딩...")
texts = load_data()
print(f"총 {len(texts)}개")

print("\n명사 추출 중... (시간이 걸릴 수 있어요)")
corpus = [extract_nouns(t) for t in texts]

print("TF-IDF Vectorizer 학습...")
vectorizer = TfidfVectorizer(max_features=50000, ngram_range=(1, 2))
vectorizer.fit(corpus)

joblib.dump({"vectorizer": vectorizer}, os.path.join(MODEL_DIR, "logreg_classifier.pkl"))
print(f"\n저장 완료 → {MODEL_DIR}/logreg_classifier.pkl")
```
